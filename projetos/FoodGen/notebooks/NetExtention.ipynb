```python
import warnings
import torch
import torchvision
from torch import nn
import numpy
```

The original ChineseFoodNet dataset seems to be unavailable, so I will train on the Food101 dataset.



```python
from torchvision.datasets import Food101
```

```python
food101set = torchvision.datasets.Food101('./data', download=False)
```

```python
batch_size = 256
transformer = torchvision.transforms.Compose([
    torchvision.transforms.Resize((64, 64)),
    torchvision.transforms.ToTensor(),
    torchvision.transforms.Normalize(0.5, 0.5)
])
food101set.transform = transformer
```

```python
from d2l import torch as d2l
data_iter = torch.utils.data.DataLoader(food101set, batch_size=batch_size, shuffle=True, num_workers=2)
d2l.set_figsize((4,4))
for X, y in data_iter:
    imgs = X[:20,:,:,:].permute(0, 2, 3, 1)/2+0.5
    d2l.show_images(imgs, num_rows=4, num_cols=5)
    break
```

```python
class G_block(nn.Module):
    def __init__(self, out_channels, in_channels=3, kernel_size=4, strides=2,
                 padding=1, **kwargs):
        super(G_block, self).__init__(**kwargs)
        self.conv2d_trans = nn.ConvTranspose2d(in_channels, out_channels,
                                kernel_size, strides, padding, bias=False)
        self.batch_norm = nn.BatchNorm2d(out_channels)
        self.activation = nn.ReLU()

    def forward(self, X):
        return self.activation(self.batch_norm(self.conv2d_trans(X)))
```

```python
num_gen_out_channels = 64
net_G = nn.Sequential(
    G_block(in_channels=100, out_channels=num_gen_out_channels*8,strides=1, padding=0), # Output: (64 * 8, 4, 4)
    G_block(in_channels=num_gen_out_channels*8, out_channels=num_gen_out_channels*4),   # Output: (64 * 4, 8, 8)
    G_block(in_channels=num_gen_out_channels*4, out_channels=num_gen_out_channels*2),   # Output: (64 * 2, 16, 16)
    G_block(in_channels=num_gen_out_channels*2, out_channels=num_gen_out_channels),     # Output: (64, 32, 32)
    nn.ConvTranspose2d(in_channels=num_gen_out_channels, out_channels=3,
                       kernel_size=4, stride=2, padding=1, bias=False),
    nn.Tanh())  # Output: (3, 64, 64)
```

```python
class D_block(nn.Module):
    def __init__(self, out_channels, in_channels=3, kernel_size=4, strides=2,
                padding=1, alpha=0.2, **kwargs):
        super(D_block, self).__init__(**kwargs)
        self.conv2d = nn.Conv2d(in_channels, out_channels, kernel_size,
                                strides, padding, bias=False)
        self.batch_norm = nn.BatchNorm2d(out_channels)
        self.activation = nn.LeakyReLU(alpha, inplace=True)

    def forward(self, X):
        return self.activation(self.batch_norm(self.conv2d(X)))
```

```python
n_dis_out_channels = 64
net_D = nn.Sequential(
    D_block(n_dis_out_channels),  # Output: (64, 32, 32)
    D_block(in_channels=n_dis_out_channels, out_channels=n_dis_out_channels*2),  # Output: (64 * 2, 16, 16)
    D_block(in_channels=n_dis_out_channels*2, out_channels=n_dis_out_channels*4),  # Output: (64 * 4, 8, 8)
    D_block(in_channels=n_dis_out_channels*4, out_channels=n_dis_out_channels*8),  # Output: (64 * 8, 4, 4)
    nn.Conv2d(in_channels=n_dis_out_channels*8, out_channels=1,
              kernel_size=4, bias=False))  # Output: (1, 1, 1)
```

```python
def train(net_D, net_G, data_iter, num_epochs, lr, latent_dim,
          device=d2l.try_gpu()):
    loss = nn.BCEWithLogitsLoss(reduction='sum')
    for w in net_D.parameters():
        nn.init.normal_(w, 0, 0.02)
    for w in net_G.parameters():
        nn.init.normal_(w, 0, 0.02)
    net_D, net_G = net_D.to(device), net_G.to(device)
    trainer_hp = {'lr': lr, 'betas': [0.5,0.99]}
    trainer_D = torch.optim.Adam(net_D.parameters(), **trainer_hp)
    trainer_G = torch.optim.Adam(net_G.parameters(), **trainer_hp)
    animator = d2l.Animator(xlabel='epoch', ylabel='loss',
                            xlim=[1, num_epochs], nrows=2, figsize=(5, 5),
                            legend=['discriminator', 'generator'])
    animator.fig.subplots_adjust(hspace=0.3)
    for epoch in range(1, num_epochs + 1):
        # Train one epoch
        print("Epoch number:", epoch)
        timer = d2l.Timer()
        metric = d2l.Accumulator(3)  # loss_D, loss_G, num_examples
        batchno = 0
        for X, _ in data_iter:
            batch_size = X.shape[0]
            Z = torch.normal(0, 1, size=(batch_size, latent_dim, 1, 1))
            X, Z = X.to(device), Z.to(device)
            #if(batchno % 3 == 0):
            metric.add(d2l.update_D(X, Z, net_D, net_G, loss, trainer_D),
                       d2l.update_G(Z, net_D, net_G, loss, trainer_G),
                       batch_size)
            #else:
            #    d2l.update_G(Z, net_D, net_G, loss, trainer_G)
                
           # batchno = batchno + 1
        # Show generated examples
        Z = torch.normal(0, 1, size=(21, latent_dim, 1, 1), device=device)
        # Normalize the synthetic data to N(0, 1)
        fake_x = net_G(Z).permute(0, 2, 3, 1) / 2 + 0.5
        imgs = torch.cat(
            [torch.cat([
                fake_x[i * 7 + j].cpu().detach() for j in range(7)], dim=1)
             for i in range(len(fake_x)//7)], dim=0)
        animator.axes[1].cla()
        animator.axes[1].imshow(imgs)
        # Show the losses
        loss_D, loss_G = metric[0] / metric[2], metric[1] / metric[2]
        animator.add(epoch, (loss_D, loss_G))
    print(f'loss_D {loss_D:.3f}, loss_G {loss_G:.3f}, '
          f'{metric[2] / timer.stop():.1f} examples/sec on {str(device)}')
```

Thus, we have all the elements set up for a nice little GAN.

```python
latent_dim, lr, num_epochs = 100, 0.0002, 100
train(net_D, net_G, data_iter, num_epochs, lr, latent_dim)
```
