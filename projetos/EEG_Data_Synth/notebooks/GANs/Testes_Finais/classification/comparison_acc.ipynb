# Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Loading data

```python
subjects = [1,3,7,9]
paths = [f"./subject_{i}/classification_results.csv" for i in subjects]
classifications_csv = [pd.read_csv(path) for path in paths ]
```

```python
[f"{int(i)}%" for i in list(classifications_csv[0]["Augmentation (%)"].values)]
```

# Plotting data

```python
# Create a figure and axis
fig, ax = plt.subplots()

# Set seaborn style for prettier plots
sns.set(style="whitegrid")

# Loop through the dataframes and plot them
for i, df in enumerate(classifications_csv):
    sns.lineplot(y='EEGNet Accuracy', x='Augmentation (%)', data=df, label=f'Subject {subjects[i]}', ax=ax)

# Set title and axis labels
ax.set_title('EEGNet Accuracy vs Augmentation for Subjects 1, 3, 7 and 9')
ax.set_ylabel('EEGNet Accuracy')
ax.set_xlabel('Augmentation (%)')

# Display legend
ax.legend()

# Show the plot
plt.show()
```

```python
# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))  # Adjust the figure size for better visibility

# Set seaborn style for prettier plots
sns.set(style="whitegrid", palette="pastel")  # Use the "husl" color palette

# Loop through the dataframes and plot them with a line plot
for i, df in enumerate(classifications_csv):
    sns.lineplot(y='EEGNet Accuracy', x='Augmentation (%)', data=df, label=f'Subject {subjects[i]}', ax=ax, marker="o", markersize=8)  # Add markers for each data point

# Set title and axis labels with a larger font size
ax.set_title('EEGNet Accuracy vs Augmentation for Subjects 1, 3, 7, and 9', fontsize=16)
ax.set_ylabel('EEGNet Accuracy', fontsize=14)
ax.set_xlabel('Augmentation (%)', fontsize=14)

# Adjust tick label font size
ax.tick_params(axis='both', labelsize=12)

# Display legend with a fancy box
ax.legend(fontsize=12, loc='lower right', bbox_to_anchor=(1.02, 0))

# Add a grid to the plot
ax.grid(True, linestyle='--', alpha=0.7)

# Remove the top and right spines for cleaner aesthetics
sns.despine()

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.savefig("EEGNet_Accuracy_vs_Augmentation.png")
plt.show()
```
